```python
from sklearn.model_selection import  train_test_split
```

```python
# Distingusih between x and y neglect -1 values 
```

```python
X=pd.DataFrame(df.iloc[:,:-1])

Y=pd.DataFrame(df.iloc[:,: -1])
```

```python
# Input 

X=df.drop('Kyphosis',axis=1)

#Output 

Y=df['Kyphosis']
```

```python
# X train and Y test 
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.3)
```

```python
print(X_train.shape)
```

```python
print(X_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier

from  sklearn import tree
```

```python
dtree=DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
#decision tree export  
```

```python
d_tree=tree.export_text(dtree)
```

```python
print(d_tree)
```

```python
df.head()
```

```python
# Tree reprsentation visualization
```

```python
fig=plt.figure(figsize=(25,20))

tree.plot_tree(dtree,feature_names=['Age','Number','Start'],class_names="Kyphosis",filled=True)
```
